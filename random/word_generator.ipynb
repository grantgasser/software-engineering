```python
import random
import numpy as np

# Define the discrete distribution of sounds
sound_dist = {'la': 2, 'ba': 1, 'baba': 2, 'opaaaa': 2, 'mama': 2, 'beebee': 2, 'woah': 2, 'wa': 2}

# Function to generate a random word based on the distribution

''.join(random.choices(list(sound_dist.keys()), weights=sound_dist.values()))
```

```python
phrase = ''
for word in range(10):
    word = ''.join(random.choices(list(sound_dist.keys()), weights=sound_dist.values(), k=np.random.randint(2, 4)))
    phrase += (word.title() + "? <break time='0s'/>")
```

```python
phrase
```

```python
from openai import OpenAI
from IPython.display import Audio

client = OpenAI()

response = client.audio.speech.create(
    model="tts-1",
    voice="shimmer",
    input=phrase
)

response.stream_to_file("ky_talk.mp3")
```

```python
# Pitch up
from pydub import AudioSegment

# Load the audio file
audio = AudioSegment.from_file("ky_talk.mp3")

# Define the amount by which you want to raise the pitch (in semitones)
pitch_shift_amount = 3  # Adjust this value as needed

# Apply pitch shift
shifted_audio = audio._spawn(audio.raw_data, overrides={
    "frame_rate": int(audio.frame_rate * (2 ** (pitch_shift_amount / 12.0)))
})

# Export the modified audio to a new file
shifted_audio.export(f"ky_talk (Pitch+{pitch_shift_amount}).wav", format="mp3")
```
